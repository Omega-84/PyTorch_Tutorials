```python
import torch
```

```python
from torch.utils.data import DataLoader
from torchvision import datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
training_data = datasets.FashionMNIST(train=True,root='./mnist',download=False,transform=transforms.ToTensor())
testing_data = datasets.FashionMNIST(train=False,root='./mnist',download=False,transform=transforms.ToTensor())
```

```python
labels_map = dict(zip(range(len(training_data.classes)),training_data.classes))
```

```python
figure = plt.figure(figsize=(8,8))
cols,rows = 3,3
for i in range( 1,cols*rows+1):
    sample_idx = torch.randint(len(training_data),size=(1,)).item()
    img,label = training_data[sample_idx]
    figure.add_subplot( rows,cols,i)
    plt.title(labels_map[label])
    plt.axis('off')
    plt.imshow(img.squeeze())
```

```python
batch_size = 64
train_loader = DataLoader(training_data,shuffle=True,batch_size=batch_size)
```

```python
img,label = next(iter(train_loader))
```

```python
plt.title(labels_map[label[10].item()])
plt.imshow(img[10].squeeze())
```
